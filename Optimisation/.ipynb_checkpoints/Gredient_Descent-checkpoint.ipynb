```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

#### Example 1 A simple cost function
 $f(x) =x^2 +x+1$, using the $ we get LaTex to show the mathematical notation

```python
def f(x):
    return x**2 +x+1
```

```python
#make data 
x_1 = np.linspace(start=-3,stop=3,num=10)
#linspace function from numpy library
# it creates a sequence of number using start value and stop valeu and number of values we need
```

```python
x_1
```

```python
# plot
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize=16)
plt.ylabel("f(x)",fontsize=16)
plt.plot(x_1,f(x_1))
plt.show()
```

if f(x) is the cost function then the lowest value is the best value

```python
# if we find the min value of x. the slope of the function equl to zero it the min value
```

## Slope & Derivative function

```python
def df(x):
    return 2*x+1
# we are going to use calculus power rule  alot
```

```python
# plot function derivate side by side
plt.figure(figsize=(10,5))
# first chart cost function
plt.subplot(2,1,1)
plt.xlim(-3,3)
plt.ylim(0,8)
plt.xlabel("x",fontsize=16)
plt.ylabel("f(x)",fontsize=16)
plt.plot(x_1,f(x_1),color ="red", linewidth=3)

plt.subplot(2,1,2)
plt.xlim(-2,3)
plt.ylim(-3,6)
plt.xlabel("x",fontsize=16)
plt.ylabel("f(x)",fontsize=16)
plt.plot(x_1,df(x_1),color ="blue", linewidth=4)
plt.grid()

plt.show()
```
