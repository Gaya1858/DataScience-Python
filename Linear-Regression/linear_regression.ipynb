```python
import pandas as pd
from pandas import DataFrame # reading CSV file and making a sheet and taking column wise data seprate
import matplotlib.pyplot as plt # for the scatter plot
from sklearn.linear_model import LinearRegression # this library to do the linear regression 
```

```python
data=pd.read_csv("cost-revenue-data.csv") # creates dataframe
print(data.describe()) # this will give below in scientific notation
X = DataFrame(data,columns=["production_budget_usd"]) #creates a pandas.core.frame.DataFrame
y = DataFrame(data,columns=["worldwide_gross_usd"])
```

```python
plt.figure(figsize =(10,6))
plt.scatter(X,y,alpha=0.25) # using matplotlib.pyplot. just give X and Y values inside
plt.title("Flim Cost vs Global Revenue")
plt.xlabel("production_budget_usd $")
plt.ylabel("worldwide_gross_usd $")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
regression = LinearRegression() # create an object for linearregression
regression.fit(X,y)
```

this below coef_ will return us the slope 

```python

regression.coef_ ### slope coeffecient 
```

intercep_ will give us the theta0 or C value

```python
regression.intercept_ ### output - movie with budget 0 is lose 7million dollar
```

```python
plt.figure(figsize =(10,6))
plt.scatter(X,y,alpha=0.25) # using matplotlib.pyplot. just give X and Y values inside
plt.plot(X,regression.predict(X),color ="orange",linewidth=4) ### this line will put fitted line on the plot
plt.title("Flim Cost vs Global Revenue")
plt.xlabel("production_budget_usd $")
plt.ylabel("worldwide_gross_usd $")
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```
